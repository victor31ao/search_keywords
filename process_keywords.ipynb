```python
import pandas as pd
```

```python
filtered_df = pd.read_csv("./n_gram_processed/filter_df.csv")
filtered_df
```

```python
compared_df = pd.read_csv("./n_gram_processed/compare_to_searchkeywords.csv")
compared_df
```

```python
from ast import literal_eval
compared_df["regex_pattern"] = compared_df["regex_pattern"].apply(literal_eval)
compared_df["matches"] = compared_df["matches"].apply(literal_eval)
```

```python
regex_pattern_df = compared_df.explode("regex_pattern")
regex_pattern_df
```

```python
expand_df = regex_pattern_df.explode("matches")
expand_df = expand_df[(expand_df["matches"].str.len() == expand_df["searchkeyword"].str.len()) & (expand_df["matches"] != expand_df["searchkeyword"])]  # 相同字數才比對、去除與searchkeyword自身一樣的match
expand_df
```

```python
filtered_count_df = filtered_df[["searchkeyword", "count"]]
filtered_count_df.columns = ["matches", "count"]

expand_df = expand_df.merge(filtered_count_df, how="inner", on="matches")
expand_df
```

```python
expand_df = expand_df[expand_df["count_x"] >= expand_df["count_y"]].reset_index(drop=True)  # 將searchkeyword數量大於matches數量的留下
expand_df
```

```python
import re
expand_df["drop"] = expand_df.apply(lambda x: False if re.compile(x["regex_pattern"]).fullmatch(x["matches"]) else True, axis=1)  # 因為"compare_to_searchkeywords.csv"的比對結果是所有pattern的總和，避免重複比較，這裡再區分一次
```

```python
columns = ['searchkeyword', 'count_x', 'regex_pattern', 'matches', 'count_y']
processed_df = expand_df[columns][expand_df["drop"] == False].copy()
# processed_df.to_csv("./n_gram_processed/processed_searchkeywords.csv", index=False, encoding="utf-8-sig")
```

```python
processed_df_n_5 = processed_df[processed_df["searchkeyword"].str.len() == 5].copy()  # 把不同字數的分開比較

# 整理出每個regex_pattern的總數
pattern_counts_n_5 = processed_df_n_5.groupby("regex_pattern", as_index=False).agg({"matches": set})
pattern_counts_n_5 = pattern_counts_n_5.explode("matches")
pattern_counts_n_5 = pattern_counts_n_5.merge(filtered_count_df, how="inner", on="matches")
pattern_counts_n_5 = pattern_counts_n_5.groupby("regex_pattern", as_index=False).agg({"count": "sum"})
pattern_counts_n_5_dict = {row["regex_pattern"]: row["count"] for _, row in pattern_counts_n_5.iterrows()}
processed_df_n_5["pattern_count"] = processed_df_n_5["regex_pattern"].map(pattern_counts_n_5_dict)
```

```python
def correct_by_dot_position(df):
    prob = df["count_y"] / df["pattern_count"]
    ratio = df["count_y"] / df["count_x"]

    if prob > 0.85:  # 高露潔牙膏,2575,高.潔牙膏,高路潔牙膏,225,230,1,高露潔牙膏  維新烏絲素,751,維.烏絲素,維生烏絲素,420,431,1,維生烏絲素  安宮牛黃丸,848,安.牛黃丸,安宫牛黃丸,101,117,1,安宫牛黃丸
        return df["searchkeyword"]

    if list(df["regex_pattern"]).index(".") == 1 or list(df["regex_pattern"]).index(".") == 2 or list(df["regex_pattern"]).index(".") == 3:
        if prob > 0.40 and ratio > 0.155:  # 高露潔牙膏,2575,高.潔牙膏,高路潔牙膏,225,230,1,高露潔牙膏  屈臣氏紙巾,345,屈臣氏.巾,屈臣氏濕巾,53,86,3,屈臣氏濕巾
            return df["matches"]
        return df["searchkeyword"]
    
    # if list(df["regex_pattern"]).index(".") == 3:
    #     if prob > 0.30 and ratio > 0.1:  # 
    #         return df["matches"]
    #     return df["searchkeyword"]
    
    if prob > 0.23 and ratio > 0.18:  # 
        return df["matches"]
    
    return df["searchkeyword"]
```

```python
# 對錯字存在的位置分別計算
processed_df_n_5["dot_position"] = processed_df_n_5["regex_pattern"].apply(lambda x: list(x).index("."))

processed_df_n_5["processed"] = processed_df_n_5.apply(lambda x: correct_by_dot_position(x), axis=1)

processed_df_n_5
```
