```python
import pandas as pd
```

```python
filtered_df = pd.read_csv("./n_gram_processed/filter_df.csv")
filtered_df
```

```python
compared_df = pd.read_csv("./n_gram_processed/compare_to_searchkeywords.csv")
compared_df
```

```python
from ast import literal_eval
compared_df["regex_pattern"] = compared_df["regex_pattern"].apply(literal_eval)
compared_df["matches"] = compared_df["matches"].apply(literal_eval)
```

```python
regex_pattern_df = compared_df.explode("regex_pattern")
regex_pattern_df
```

```python
(regex_pattern_df["regex_pattern"].value_counts() > 1).value_counts()
```

```python
regex_pattern_df.drop_duplicates("regex_pattern", inplace=True)
regex_pattern_df = regex_pattern_df.reset_index(drop=True)
(regex_pattern_df["regex_pattern"].value_counts() > 1).value_counts()
```

```python
expand_df = regex_pattern_df.explode("matches")
expand_df = expand_df[(expand_df["matches"].str.len() == expand_df["searchkeyword"].str.len()) & (expand_df["matches"] != expand_df["searchkeyword"])]
expand_df
```

```python
filtered_count_df = filtered_df[["searchkeyword", "count"]]
filtered_count_df.columns = ["matches", "count"]

expand_df = expand_df.merge(filtered_count_df, how="inner", on="matches")
expand_df
```

```python
expand_df = expand_df[expand_df["count_x"] >= expand_df["count_y"]].reset_index(drop=True)
expand_df
```

```python
import re
expand_df["drop"] = expand_df.apply(lambda x: False if re.compile(x["regex_pattern"]).fullmatch(x["matches"]) else True, axis=1)
```

```python
columns = ['searchkeyword', 'count_x', 'regex_pattern', 'matches', 'count_y']
processed_df = expand_df[columns][expand_df["drop"] == False].copy()
```

```python
processed_df["processed"] = processed_df.apply(lambda x: x["searchkeyword"] if (x["count_y"]/x["count_x"]<0.16) or (x["count_y"]<50) else x["matches"], axis=1)
processed_df
```

```python
processed_df.drop_duplicates(["searchkeyword", "matches"], inplace=True)
processed_df = processed_df.reset_index(drop=True)
processed_df
```

```python
processed_df.to_csv("./n_gram_processed/processed_searchkeywords.csv", index=False, encoding="utf-8-sig")
processed_df[["matches", "processed"]].to_csv("./n_gram_processed/processed_searchkeywords_result.csv", index=False, encoding="utf-8-sig")
```
